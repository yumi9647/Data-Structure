```python
## 괄호로 묶인 수식으로 가정 


class Calculator():
    def __init__(self):
        self.equation1 = [] #입력받은 수식 리스트
        self.equation2 = [] #후위표기법 스택
        self.calculators = [] # 연산자 스택
        self.output = []
        self.num = 0
        self.top = -1 #연산자 top
        self.top2 = -1 # equation2 top
        self.top3 = -1 # output top
    
    def strToList(self, sentence): #입력받은 수식을 리스트로 만들기
        for i in sentence:
            if (i == "(") or (i == ")") or (i == "+") or (i == "-")or (i == "*") or (i == "/"):
                i = [i]
                self.equation1 = self.equation1 + i
            else:
                i = int(i)
                i = [i] ###숫자를 리스트로 
                self.equation1 = self.equation1 + i
    
    def createDummy(self): #입력받은 수식만큼 연산자 스택과 후위표기법 스택에 dummy 데이터 생성
        self.num = len(self.equation1)
        for i in range(self.num):
            dummy = ['x']
            self.calculators = self.calculators + dummy
        for i in range(self.num):
            dummy = ['x']
            self.equation2 = self.equation2 + dummy
        for i in range(self.num):
            dummy = ['x']
            self.output = self.output + dummy
        
    def middleToLatter(self): # 중위를 후위 식으로 변환
        for i in self.equation1:
            if i == "(": #왼쪽 괄호 무시
                pass
            elif type(i) == int: # 숫자이면 후위표기법 스택에 저장
                self.top2 += 1
                self.equation2[self.top2] = i
            elif (i == "+") or (i == "-")or (i == "*") or (i == "/"): #연산자는 스택에 저장
                self.top = self.top + 1
                self.calculators[self.top] = i
            elif i == ")":
                self.top2 += 1
                self.equation2[self.top2] = self.calculators[self.top] #후위표기법 스택에 연산자 스택의 가장 최신 값을 저장
                self.calculators[self.top] = 'x' #연산자 스택 delete
                self.top = self.top -1
        if self.top != -1:
            for i in range(self.top+1):
                self.top2 += 1
                self.equation2[self.top2] = self.calculators[self.top] #후위표기법 스택에 연산자 스택의 가장 최신 값을 저장
                self.calculators[self.top] = 'x' #연산자 스택 delete
                self.top = self.top -1
    
    def Output(self):
        for i in self.equation2:
            if type(i) == int:
                self.top3 += 1
                self.output[self.top3] = i
            elif (i == "+"):
                b = self.output[self.top3]
                a = self.output[self.top3 -1]
                outcome = a + b
                self.output[self.top3] = 'x'
                self.output[self.top3 -1] = 'x'
                self.top3 = self.top3 -1
                self.output[self.top3] = outcome
            elif (i == "-"):
                b = self.output[self.top3]
                a = self.output[self.top3 -1]
                outcome = a - b
                self.output[self.top3] = 'x'
                self.output[self.top3 -1] = 'x'
                self.top3 = self.top3 -1
                self.output[self.top3] = outcome
            elif (i == "*"):
                b = self.output[self.top3]
                a = self.output[self.top3 -1]
                outcome = a * b
                self.output[self.top3] = 'x'
                self.output[self.top3 -1] = 'x'
                self.top3 = self.top3 -1
                self.output[self.top3] = outcome
            elif (i == "/"):
                b = self.output[self.top3]
                a = self.output[self.top3 -1]
                outcome = a / b
                self.output[self.top3] = 'x'
                self.output[self.top3 -1] = 'x'
                self.top3 = self.top3 -1
                self.output[self.top3] = outcome
        return self.output[self.top3]

        
```
