```python
class Heap: #최대 힙
    def __init__(self):
        self.tree = ['x']
        self.numData = len(self.tree)-1
    
    def isEmpty(self):
        if self.numData == 0:
            return True
        else:
            return False
    
    def insert(self, data):                  #숙제숙제  
        if self.isEmpty() == True:
            self.tree.append(data)
        elif self.isEmpty() == False:
            self.tree.append(data)
            childIndex = len(self.tree)-1
            while True:
                parentIndex = childIndex//2
                if childIndex is 1: ## < 1
                    break
                elif self.tree[parentIndex] > self.tree[childIndex]:
                    break
                else:
                    # 자리 바꾸기, child 값 업데이트
                    temp = self.tree[parentIndex]
                    self.tree[parentIndex] = self.tree[childIndex]
                    self.tree[childIndex] = temp 
                    childIndex = parentIndex
        self.numData = self.numData + 1
        
    def delete(self):
        if self.isEmpty == True:
            pass
        else:
            self.tree.pop(1)  #루트 제거 -> 삭제는 루트만 제거하기 때문
            currentIndex = 1
            while True:
                if currentIndex*2 > len(self.tree)-1: #트리 끝단에 도달
                    break
                elif self.tree[currentIndex] > self.tree[currentIndex*2] and self.tree[currentIndex]> self.tree[currentIndex*2+1]:
                    break  #왼쪽자식 / 오른쪽 자식보다 클 때 
                else: #자식 중 큰 값과 자리 교환
                    first_child = currentIndex*2
                    second_child = currentIndex*2+1
                    if self.tree[first_child] > self.tree[second_child]:
                        temp = self.tree[currentIndex]
                        self.tree[currentIndex] = self.tree[first_child]
                        self.tree[first_child] = temp
                        currentIndex = first_child # 인덱스 값 업데이트 
                    elif self.tree[second_child] > self.tree[first_child]:
                        temp = self.tree[currentIndex]
                        self.tree[currentIndex] = self.tree[second_child]
                        self.tree[second_child] = temp
                        currentIndex = second_child
                    
            
            
```
